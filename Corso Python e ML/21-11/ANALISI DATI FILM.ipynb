```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = r"C:\Users\bugli\Desktop\GitHub\Corso-PY-e-ML\Corso Python e ML\Dataset\TMDB  IMDB Movies Dataset.csv"
df = pd.read_csv(file_path)


df_pulito=df.dropna(subset=['budget','revenue']) #revenue=incassi
df_pulito=df_pulito[df_pulito['runtime'] >0]


correlazione_voto_tempo = df_pulito[['runtime', 'vote_average']].corr()
print(f"correlazione tra tempo e voto medio: \n{correlazione_voto_tempo}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='runtime', y='vote_average', data=df_pulito, alpha=0.5)
plt.title('Correlazione tra durata del film e voto medio', fontsize=16)
plt.xlabel('Durata del film (minuti)', fontsize=12)
plt.ylabel('Voto medio dell\'utente', fontsize=12)
plt.grid(True)
plt.show()
```

```python
df_filtrato = df_pulito[df_pulito['runtime'] <= 2000]
correlazione_voto_tempo_filtrato = df_filtrato[['runtime', 'vote_average']].corr()

print(f"Correlazione tra durata del film e voto medio (film fino a 2000 minuti): \n{correlazione_voto_tempo_filtrato}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='runtime', y='vote_average', data=df_filtrato, alpha=0.5)
plt.title('Correlazione tra durata del film e voto medio (fino a 2000 minuti)', fontsize=16)
plt.xlabel('Durata del film (minuti)', fontsize=12)
plt.ylabel('Voto medio dell\'utente', fontsize=12)
plt.grid(True)
plt.show()
```

```python
df_filtrato_60_2000 = df_pulito[(df_pulito['runtime'] >= 60) & (df_pulito['runtime'] <= 2000)]

correlazione_voto_tempo_60_2000 = df_filtrato_60_2000[['runtime', 'vote_average']].corr()
print(f"Correlazione tra durata del film e voto medio (film tra 60 e 2000 minuti): \n{correlazione_voto_tempo_60_2000}")
```
